# Data Collection and Source Documentation


## Purpose

This notebook documents the raw data sources used in the AI Infrastructure Economics project. Most values are currently stored manually in CSV files under `/sources`. This notebook exists to make the data pipeline transparent and reproducible.


## Datasets Used

| Dataset | Variables | Source | Stored In | Notes |
| --- | --- | --- | --- | --- |
| EIA State Electricity Profiles | Electricity price; installed generation capacity; renewable share | U.S. Energy Information Administration | `sources/state_competitiveness.csv` | Core grid-cost and power-system inputs for the competitiveness model. |
| DataCenterMap | Data center count / ecosystem proxy | DataCenterMap | `sources/state_competitiveness.csv` | Used to construct the data center ecosystem score. |
| Interconnection / Queue Friction Sources | Interconnection region; interconnection friction score | Interconnection and queue-friction literature and ISO/RTO documentation | `sources/state_competitiveness.csv` | Regional proxy based on interconnection-region friction, not direct state-level queue duration. |
| EIA Natural Gas State Data | Natural gas price | U.S. Energy Information Administration | `sources/btm_inputs.csv` | Used for behind-the-meter gas generation fuel-cost inputs. |
| NREL PVWatts | Solar capacity factor | NREL PVWatts | `sources/btm_inputs.csv` | Derived as Annual AC Energy / (System Size × 8760). |
