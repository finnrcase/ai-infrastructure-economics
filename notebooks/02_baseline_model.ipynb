# Baseline Model Development

## Purpose

This notebook documents the development process behind the AI Infrastructure Competitiveness Model.

The goal was to identify a small set of publicly available variables capable of approximating the attractiveness of different U.S. states for future AI infrastructure deployment.

Unlike the final model in Notebook 3, this notebook focuses on research design, variable selection, normalization choices, and model construction decisions.

---

## Research Question

How can AI infrastructure competitiveness be approximated using publicly available data on electricity markets, power systems, and existing infrastructure ecosystems?

More specifically:

How will electricity prices, power availability, infrastructure ecosystems, and grid constraints influence future AI infrastructure deployment across the United States?

---

## Initial Modeling Philosophy

The objective was not to predict exactly where future data centers will be built.

Instead, the model seeks to create a comparative framework that captures the relative attractiveness of states under different infrastructure conditions.

The model prioritizes:

- Simplicity
- Interpretability
- Publicly available data
- Infrastructure relevance
- Reproducibility

---

## Candidate Variables Considered

### Variables Implemented in Version 1

- Electricity Price
- Grid Capacity
- Renewable Share
- Existing Data Center Ecosystem
- Behind-the-Meter Potential
- Interconnection Friction

### Variables Considered but Not Implemented

- Carbon Intensity
- Water Stress
- Land Availability
- Transmission Congestion
- Reliability Metrics
- Nuclear Capacity
- Geothermal Potential
- State-Level Queue Duration
- Load Growth Pressure

These variables were excluded from Version 1 due to limited state-level availability, inconsistent reporting, or concerns about introducing excessive model complexity.

---

## Normalization Approach

Because variables exist in different units, all variables are transformed into normalized scores.

Examples:

### Electricity Price

Lower electricity prices improve competitiveness.

Price Score:

price_score = 1 − normalized_price

### Grid Capacity

States with larger installed generation fleets receive higher scores.

Capacity Score:

capacity_score = normalized_capacity

### Renewable Share

Higher renewable penetration improves competitiveness.

Renewable Score:

renewable_score = normalized_renewable_share

---

## Weighting Strategy

Several weighting approaches were considered:

### Equal Weight Model

All variables receive identical importance.

Advantages:
- Simple
- Transparent

Disadvantages:
- Assumes all variables are equally important

### Price-Heavy Model

Prioritizes electricity economics.

Advantages:
- Reflects hyperscaler focus on power cost

Disadvantages:
- May understate ecosystem effects

### Capacity-Heavy Model

Prioritizes power availability.

Advantages:
- Reflects emerging grid constraints

Disadvantages:
- May understate economics

---

## Why Monte Carlo Was Added

Fixed weights introduce subjective assumptions.

To reduce this issue, Notebook 4 introduces Monte Carlo simulations that randomly generate thousands of weight combinations.

This allows the analysis to evaluate:

- Ranking robustness
- Winner frequency
- Sensitivity to modeling assumptions
- State performance across many possible futures

Rather than asking:

"Who wins under one set of weights?"

the project asks:

"Who wins most often across many plausible weighting schemes?"

---

## Interconnection Friction

The original model did not include transmission interconnection constraints.

Version 1.1 introduces an Interconnection Friction Score based on major U.S. interconnection regions.

The score acts as a proxy for permitting and queue difficulty.

This approach was chosen because reliable state-level queue duration data was not consistently available.
